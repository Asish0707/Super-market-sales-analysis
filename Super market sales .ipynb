```python
import matplotlib.pyplot as plt
import numpy as np 
import seaborn as sns
import pandas as pd 

np.random.seed(42)
n=500
data={
    'Invoice Id':[f'Inv{1000+i}' for i in range (n)],
    'Customer_type':np.random.choice(['member','normal'],n),
    'Gender':np.random.choice(['male','female'],n),
    'Product_line':np.random.choice(['Beverages','Beakery','Eloctronics','clothing','home decor','Health'],n),
    'Unit_price':np.random.uniform(5,200,n).round(2),
    'Quantity':np.random.randint(1,10,n),
    'Payment':np.random.choice(['cash','UPI','card'],n),
    'Hours':np.random.randint(9,22,n),
    'Rating':np.random.randint(1,5,n),
}

df=pd.DataFrame(data)

df['Tax']=((df['Unit_price']*df['Quantity'])*0.05).round(2)

df['Total']=(df['Unit_price']*df['Quantity'])+df['Tax'].round(2)

print(df)

df.to_csv('SuperMarket_sales.csv',index=False)
df


    

```

```python
# visualisation

plt.figure(figsize=(8,4))
sns.barplot(x='Product_line',y='Total',data=df,estimator=sum)
plt.title('Revenue by product line')
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.barplot(x='Gender',y='Total',data=df)
plt.title('Average by gender')
plt.show()
```

```python
plt.figure(figsize=(6,6))
df['Customer_type'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Customer type distribution')
plt.ylabel("")
plt.show()
```

```python
# hourly sales trend
plt.figure(figsize=(8,5))
sns.lineplot(x="Hours", y="Total", data=df, estimator=sum)
plt.title("Total Sales by Hour")
plt.show()

```

```python
plt.figure(figsize=(7,7))
sns.heatmap(df.corr(numeric_only=True),annot=True,cmap='coolwarm')
plt.title('Hetmap')
plt.show()
            
```

```python
#Distribution of total bill

plt.figure(figsize=(8,6))
sns.histplot(x='Total',data=df,kde=True)
plt.title("Distribution by ")
plt.show()
```
